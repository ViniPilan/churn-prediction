```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.axes as axes
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
ax = sns.barplot(y=['N° de clientes\nque saíram', 'N° total de\nclientes'], x=[25, 100], palette='Oranges_r')
plt.axis('off')

ax.text(x=30, y=-0.1, s='25% dos clientes totais', size=16, color=colors.to_rgb((189/255, 102/255, 26/255)), fontweight='bold')
ax.text(x=30, y=0.1, s='saem da empresa por ano', size=16, 
        color=colors.to_rgb((189/255, 102/255, 26/255)), fontweight='normal')
ax.text(x=10, y=1.6, s='Total de clientes da empresa', size=16, color=colors.to_rgb((191/255, 141/255, 98/255)), fontweight='bold')


plt.savefig('../Images/estimativa_saida.jpg', bbox_inches="tight")
plt.show()
```
